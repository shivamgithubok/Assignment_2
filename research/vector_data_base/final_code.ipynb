````python
import pytesseract
from pdf2image import convert_from_path

# List of PDF files
pdf_files = [
    "pdf_1.pdf", "pdf_2.pdf", "pdf_3.pdf", "pdf_4.pdf", "pdf_5.pdf",
    "pdf_6.pdf", "pdf_7.pdf", "pdf_8.pdf", "pdf_9.pdf", "pdf_10.pdf",
    "pdf_11.pdf", "pdf_12.pdf", "pdf_13.pdf", "pdf_14.pdf", "pdf_15.pdf",
    "pdf_16.pdf", "pdf_17.pdf", "pdf_18.pdf", "pdf_19.pdf", "pdf_20.pdf",
    "pdf_21.pdf", "pdf_22.pdf", "pdf_23.pdf", "pdf_24.pdf", "pdf_25.pdf"
]

output_txt = "extracted_text.txt"

# Open the output file once to store all text
with open(output_txt, "w", encoding="utf-8") as file:
    for pdf_path in pdf_files:  # Process each PDF file separately
        try:
            images = convert_from_path(pdf_path)  # Convert PDF to images
            file.write(f"=== Extracted text from {pdf_path} ===\n")

            for i, img in enumerate(images):
                text = pytesseract.image_to_string(img)
                file.write(f"--- Page {i+1} ---\n")
                file.write(text + "\n\n")

            print(f"✅ OCR completed for {pdf_path}")

        except Exception as e:
            print(f"❌ Error processing {pdf_path}: {e}")

print(f"✅ All PDFs processed! Extracted text saved to '{output_txt}'")
````

````python
import re

# List of all scraped text files
file_paths = ["segment_all_docs.txt", "extracted_text.txt", "Lytics_all_docs.txt"]

# Read and combine all files
data = ""
for file_path in file_paths:
    with open(file_path, "r", encoding="utf-8") as file:
        data += file.read() + "\n"  # Adding newline for separation

# 1. Remove HTML Tags
data = re.sub(r"<.*?>", "", data)

# 2. Remove Special Characters (except common punctuation)
data = re.sub(r"[^a-zA-Z0-9,.!?;:\s]", "", data)

# 3. Remove Extra Whitespaces & Newlines
data = re.sub(r"\s+", " ", data).strip()

# 4. Convert to Lowercase
data = data.lower()

# Save the cleaned text
with open("cleaned_data.txt", "w", encoding="utf-8") as file:
    file.write(data)

print("✅ Data cleaning completed! Saved as 'cleaned_data.txt'")
````

````python
import json
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

# Load the JSON data
with open("cleaned_data.json", "r", encoding="utf-8") as file:
    data = json.load(file)

# Extract text data from JSON
documents = data["CDP_Documentation"].split(". ")  # Split into sentences

# Load a pre-trained sentence transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")  # Small, fast model

# Convert text documents to embeddings
document_embeddings = model.encode(documents, convert_to_numpy=True)

# Create FAISS index (L2 normalization improves search)
dimension = document_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)  # L2 distance for similarity
index.add(document_embeddings)  # Add embeddings to FAISS index

# Save FAISS index
faiss.write_index(index, "faiss_index.bin")
np.save("documents.npy", np.array(documents))

print("✅ FAISS index built and saved!")
````

````python
from fastapi import FastAPI
from pydantic import BaseModel
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

# Load FAISS index and documents
index = faiss.read_index("faiss_index.bin")
documents = np.load("documents.npy", allow_pickle=True)

# Load the sentence transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")

app = FastAPI()

class QueryRequest(BaseModel):
    question: str

# Function to search FAISS
def search_faiss(query, top_n=5):
    query_embedding = model.encode([query], convert_to_numpy=True)
    _, indices = index.search(query_embedding, top_n)
    return [documents[i] for i in indices[0]]

@app.post("/ask")
def ask_question(request: QueryRequest):
    results = search_faiss(request.question)
    return {"answers": results}

# Run the server using: uvicorn filename:app --reload
````
